```python
def particionar(lista, inicio, final):
    x = lista[final]
    i = inicio -1
    list_izq = []
    list_der = []
    list_copy = lista.copy()
    for j in range(inicio, final):
        print(f"lista {list_izq} {list_der} {list_copy}")
        if lista[j] <= x:
            list_izq.append(lista[j])
            list_copy.remove(lista[j])
            i += 1
            lista[i], lista[j] = lista[j], lista[i]
        else:
            list_copy.remove(lista[j])
            list_der.append(lista[j])
    print(f"lista {list_izq} {list_der} {list_copy}")
    lista[i+1], lista[final] = lista[final], lista[i+1]
    list_der = lista[i+2:final+1]
    print(f"lista {list_izq} {x} {list_der}")
    return i+1

if __name__ == "__main__":
    lista = [2,8,7,1,3,5,6,4]
    print(lista)
    i = particionar(lista, 0, len(lista)-1)
    print(lista)
```

```python
import math

def hIzq(i):
    return 2*i+1

def hDer(i):
    return 2*i+2

def MaxHeapify(heap, i, tamanioHeap):
    L = hIzq(i)
    R = hDer(i)
    if (L <= tamanioHeap-1) and (heap[L] > heap[i]):
        posMax = L
    else:
        posMax = i
    if (R <= tamanioHeap-1) and (heap[R] > heap[posMax]):
        posMax = R
    if posMax != i:
        heap[i], heap[posMax] = heap[posMax], heap[i]
        MaxHeapify(heap, posMax, tamanioHeap)

def construirMaxIni(heap, tamanioHeap):
    print(f"Heap maximo inicial\n {heap}\nEmpieza ordenamiento")
    for i in range(math.ceil((tamanioHeap-1)/2), -1, -1):
        MaxHeapify(heap, i, tamanioHeap)

def OrdenacionHeapSort(heap, tamanioHeap):
    heap_izq = []
    heap_der = []
    construirMaxIni(heap, tamanioHeap)
    for i in range(len(heap)-1, 0, -1):
        print(f"\nIntercambio de A[0] = {heap[0]} con A[{i}] = {heap[i]}\n", 
                "Despues de intercambio no es heap max")
        heap[0], heap[i] = heap[i], heap[0]
        heap_izq = heap[:i]
        heap_der = heap[i:]
        print(heap_izq, heap_der)
        tamanioHeap -= 1
        MaxHeapify(heap, 0, tamanioHeap)
        print("\nReconstruccion del heap")
        print(heap[:i], heap_der)

if __name__ == "__main__":
    heap = [16,14,10,8,7,9,3,2,4,1,18,13,11,5,12,17,15,19,6,20]
    OrdenacionHeapSort(heap, len(heap))
    print(heap)
```
